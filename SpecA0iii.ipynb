```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('uka0iii.fits')    #4771 values
hdu = fits.open('uka0iii.fits')
hdu[0].header   
#print (dd)
```

```python
c = hdu[0].data[:]
#len(c)
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type A0iii')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,c,'b');plt.show()
```

```python
plt.xlim(3500,10000)
plt.plot(wl,c,'b');plt.show()
```
