```python
#spec-0280-51612-0549
### SpecObjID = 315402908931745792
### http://skyserver.sdss.org/dr12/en/tools/explore/Summary.aspx?id=1237648722300567644

import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('spec-0280-51612-0549.fits')    #(3840R,8C)
datad = fits.open('spec-0280-51612-0549.fits')
dd = datad[0].header   #central wavelength = 3.5789, 0.0001/pix. (log10)

print (dd)
```

```python
#c = datad[1].data.names
#print (c)
specd = datad[1].data.field('flux')   
# or specd = data[1].data.field(0)
#print (specd)
wld = np.linspace(3.5789,3.9629,3840)
lgd = np.power(10,wld)

plt.figure(figsize=(30, 20))
plt.title('spec-0280-51612-0549')
plt.xlabel('wavelength')
plt.ylabel('flux')

plt.plot(lgd,specd,'b');plt.show()
```
