```python
#!/usr/bin/env python
### SpecObjID = 396430595082708992
### http://skyserver.sdss.org/dr12/en/tools/explore/Summary.aspx?id=1237671939804626959

import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('spec-0352-51694-0414.fits')    #(3579R,8C)
data = fits.open('spec-0352-51694-0414.fits')
dd = data[0].header   #central wavelength = 3.579, 0.0001/pix. (log10)

#data.close()
#print (dd)
```

```python
c = data[1].data.names
#print (c)
spec = data[1].data.field('flux')   
# or spec = data[1].data.field(0)
#print (spec)
wla = np.linspace(3.5790,3.9630,3840)
lga = np.power(10,wla)

plt.figure(figsize=(30, 20))
plt.title('spec-0352-51694-0414')
plt.xlabel('wavelength')
plt.ylabel('flux')

plt.plot(lga,spec,'b');plt.show()
```
