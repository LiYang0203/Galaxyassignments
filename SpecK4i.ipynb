```python
#ukk4i
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('ukk4i.fits')    
hdu = fits.open('ukk4i.fits')
dd = hdu[0].header   # 4771 values, with beginning of 1150?
print (dd)
```

```python
ck = hdu[0].data[:]
#len(ck)
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type K4i')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,ck,'b');plt.show()
```

```python
plt.xlim(3500,10000)
plt.plot(wl,ck,'b');plt.show()
```
