```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('ukf8i.fits')    #4771 values
datad = fits.open('ukf8i.fits')
dd = datad[0].header   #start with 1150 ?
print (dd)
```

```python
c = datad[0].data[:]
len(c)
```

```python
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type F8i')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,c,'b');plt.show()
```

```python
plt.xlim(3000,10000)
plt.plot(wl,c,'b');plt.show()
```
