```python
#!/usr/bin/env python
### SpecObjID = 394099630381492224
### http://skyserver.sdss.org/dr12/en/tools/explore/Summary.aspx?id=1237671939804561999

import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('spec-0350-51691-0126.fits')    #(3844R,8C)
datac = fits.open('spec-0350-51691-0126.fits')
#dd = datac[0].header   #central wavelength = 3.579, 0.0001/pix. (log10)

#print (dd)
```

```python
#c = datac[1].data.names
#print (c)
specc = datac[1].data.field('flux')   
# or spec = data[1].data.field(0)
#print (specc)
wlc = np.linspace(3.5790,3.9634,3844)
lgc = np.power(10,wlc)

plt.figure(figsize=(30, 20))
plt.title('spec-0350-51691-0126')
plt.xlabel('wavelength')
plt.ylabel('flux')

plt.plot(lgc,specc,'b');plt.show()
```
