```python
#!/usr/bin/env python
from astropy.io import fits
import numpy as np
#fits.info('Hipp.fit')
dat = fits.open('Hipp.fit')
Vmag = dat[1].data.field(3)
#VV = np.array(Vmag)
p = dat[1].data.field(6)	
pp = p/1000
BV = dat[1].data.field(10)
#PL=np.array([[Vmag],[pp],[BV]])
PL = np.array([Vmag,pp,BV],dtype=float)
PL = np.transpose(PL)

sel = PL[:,1] > 0.0
#print(shape(sel))
PL=PL[sel,:]


#for i in PL[sel]:
#   print(i[0],j[1],k[2])


#mag = Vmag+5+5*np.log10(pp)
#BV = dat[1].data.field(10)
#A = PL[:,PL[1]!=0]
a1=PL[:,0] #Vmag
a2=PL[:,1] #pp
a3=PL[:,2] #BV

```
