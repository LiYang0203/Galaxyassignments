```python
#ukb1i
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('ukb1i.fits')  
hdu = fits.open('ukb1i.fits')
dd = hdu[0].header   # 4771 values, with beginning of 1150?
print (dd)
```

```python
cb = hdu[0].data[:]
#len(cb)
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type B1i')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,cb,'b');plt.show()
```

```python
plt.xlim(3500,9500)
plt.ylim(0,6)
plt.plot(wl,cb,'b');plt.show()
```
