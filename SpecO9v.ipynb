```python
#uko9v
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('uko9v.fits')    
hdu = fits.open('uko9v.fits')
dd = hdu[0].header   # 4771 values, with beginning of 1150?
print (dd)
```

```python
co = hdu[0].data[:]
#len(co)
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type O9v')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,co,'b');plt.show()
```

```python
plt.xlim(3500,9500)
plt.ylim(0,6.5)
plt.plot(wl,co,'b');plt.show()
```
