```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
#from SpecK4i.py import ck
#from SpecO9v import co
#from SpecB1i import cb

fits.info('ukb1i.fits')  
hdub = fits.open('ukb1i.fits')
cb = hdub[0].data[:]

fits.info('ukk4i.fits')  
hduk = fits.open('ukk4i.fits')
ck = hduk[0].data[:]

fits.info('ukm1v.fits')  
hdum = fits.open('ukm1v.fits')
cm = hdum[0].data[:]

fits.info('xxxx-ukrf6v.fits')    #4771 values
rf6v = fits.open('xxxx-ukrf6v.fits')
rf = rf6v[0].data[:]
```

```python
C = cb + 30*ck + 20*cm + 50*rf
C2 = cb + 30*ck + 20*cm
```

```python
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of a "red" galaxy')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.xlim(3500,9500)
plt.plot(wl,C,'r',wl,C2,'b');plt.show()
```
