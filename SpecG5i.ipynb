```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

fits.info('ukg5i.fits')    
hdu = fits.open('ukg5i.fits')
dd = hdu[0].header   # 4771 values, with beginning of 1150?
print (dd)
```

```python
c = hdu[0].data[:]
#len(c)
wl = np.linspace(1150,25000,4771)
plt.title('Spectrum of star type G5i')
plt.xlabel('Wavelength')
plt.ylabel('Flux(lambda)')
plt.plot(wl,c,'b');plt.show()
```

```python
plt.xlim(3500,10000)
plt.plot(wl,c,'b');plt.show()
```
